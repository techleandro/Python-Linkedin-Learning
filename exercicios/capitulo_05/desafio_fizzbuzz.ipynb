- Crie um objeto iterável com 100 dados do tipo numérico e imprima na tela a palavra 'Fizz' no lugar dos números divisíveis por 3, 'Buzz' no lugar dos números divisíveis por 05 e 'FizBuzz' no lugar dos números divisíveis por 3 e 5. Caso o número não atenda a nenhuma dessas condições, imprima o próprio número.

```python
objeto = range(0,101)

Fizz = []
Buzz = []
FizzBuzz = []
```

```python
for i in objeto: print(i) if i == 0 else print('FizzBuzz') if i % 3 == 0 and i % 5 == 0 else print('Fizz') if i % 3 == 0 else print('Buzz') if i % 5 == 0 else print(i)
```

```python
for i in objeto:
    if i % 3 == 0 and i % 5 == 0:
        if i != 0:
            print('FizzBuzz')
            FizzBuzz.append(i)
        else:
            print(i)
    elif i % 3 == 0:
        print('Fizz')
        Fizz.append(i)
    elif i % 5 == 0:
        print('Buzz')
        Buzz.append(i)
    else:
        print(i)
```
